f(x) = (sin(x)/2 + cos(x)/2) ** 2 * sin(x)  

    


```python
import math
from sympy import *
x = symbols('x', real=True)
y = (sin(x)/2 + cos(x)/2) ** 2 * sin(x)
y
```

Определить корни

```python
roots = solve(y)
roots
```

Построить график

```python
graf = plot(y)
graf
```

Вычислить вершину

```python
fd=diff(y)
corni=solve(fd)
top=corni[0]
x=top
y=(sin(x)/2 + cos(x)/2) ** 2 * sin(x)
print(top,y)
```

Определить промежутки, на котором f > 0

```python
solve(0 < y)
```

Определить промежутки, на котором f < 0

```python
solve(y < 0)
```
